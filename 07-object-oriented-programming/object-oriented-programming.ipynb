## Problem 1: Smart Light Controller

**Problem Statement:** Create a class to control a smart light that can be turned ON or OFF and display its status.

**Real-Time Use:**
- Smart home automation systems

**Sample Input:**
Light Name: Bedroom Light  
Action: ON  

**Expected Output:**
Bedroom Light is ON

```python
# Class for Smart Light
class SmartLight:
    
    # Constructor to set light name
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display current status
    def display_status(self):
        print(self.name, "is", self.status)

light = SmartLight("Bedroom Light")

light.turn_on()
light.display_status()
```

## Problem 2: Employee ID Card System

**Problem Statement:** Create a class that stores employee name, ID, and department and displays ID card details.

**Real-Time Use:**
- Corporate HR systems

**Sample Input:**
Employee Name: Rahul  
Employee ID: EMP102  
Department: IT  

**Expected Output:**
Employee ID Card  
Name: Rahul  
ID: EMP102  
Department: IT

```python
# Class for Employee details
class Employee:
    
    # Constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

employee1 = Employee("Max", "EMP3", "IT")
employee1.display_id_card()
```

## Problem 3: Mobile Contact Record

**Problem Statement:** Create a class to store contact name and phone number and display contact information.

**Real-Time Use:**
- Mobile contact applications

**Sample Input:**
Contact Name: Anita  
Phone Number: 9876543210  

**Expected Output:**
Contact Saved  
Name: Anita  
Phone: 9876543210

```python
# Class for mobile contact
class Contact:
    
    # Constructor to initialize contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

contact1 = Contact("Max", "9876543210")
contact1.display_contact()
```

## Problem 4: Product Price Tag Generator

**Problem Statement:** Create a class that stores product name and price and prints a formatted price tag.

**Real-Time Use:**
- Retail billing systems

**Sample Input:**
Product Name: Headphones  
Price: 2499  

**Expected Output:**
Product: Headphones  
Price: ₹2499

```python
# Class to represent a product
class Product:
    
    # Constructor to initialize product details
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    # Method to print formatted price tag
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

product1 = Product("Headphones", 2499)
product1.print_price_tag()
```

## Problem 5: Movie Rating Display System

**Problem Statement:** Create a class that stores movie name and rating and displays movie details.

**Real-Time Use:**
- Streaming platforms

**Sample Input:**
Movie Name: Inception  
Rating: 4.8  

**Expected Output:**
Movie: Inception  
Rating: 4.8 / 5

```python
# Class to represent a movie
class Movie:
    
    # Constructor to initialize movie details
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    # Method to display movie information
    def display_details(self):
        print("Movie:", self.name)
        print("Rating:", str(self.rating), "/ 5")

movie1 = Movie("F1 The Movie", 4.9)
movie1.display_details()
```

## Problem 6: Delivery Address Manager

**Problem Statement:** Create a class that stores customer name and delivery address and prints delivery details.

**Real-Time Use:**
- Courier and logistics platforms

**Sample Input:**
Customer Name: Suman  
Address: Hyderabad  

**Expected Output:**
Delivery Details  
Customer: Suman  
Address: Hyderabad

```python
# Class to manage delivery details
class Delivery:
    
    # Constructor to initialize customer details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

delivery1 = Delivery("Max", "Dutch")
delivery1.display_delivery()
```
