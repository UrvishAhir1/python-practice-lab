## Problem Statement 1: Smart Parking Lot Management System
Design a function to manage a smart parking lot.

**The system should:**
- Accept vehicle entry and exit logs
- Calculate total parked vehicles
- Identify peak parking usage
- Alert if parking exceeds capacity

**Real-Time Use:**
- Mall parking systems
- Smart city infrastructure

**Hint:**
- Function + list
- Loop for counting
- Conditional alert

**Sample Input:**
Parking Capacity: 50  
Vehicle Logs: ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

**Expected Output:**
Currently Parked Vehicles: 3  
Parking Status: Available

```python
# Function to manage parking system
def manage_parking(capacity, logs):
    
    current_parked = 0
    peak_usage = 0
    
    # Process each log entry
    for entry in logs:
        
        if entry == "IN":
            current_parked += 1
        elif entry == "OUT":
            current_parked -= 1
        
        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked
    
    # Check parking status
    if current_parked > capacity:
        status = "Parking Full"
    else:
        status = "Available"
    
    return current_parked, peak_usage, status

parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parked, peak, parking_status = manage_parking(parking_capacity, vehicle_logs)

print("Currently Parked Vehicles:", parked)
print("Peak Parking Usage:", peak)
print("Parking Status:", parking_status)
```

## Problem Statement 2: Online Food Delivery Time Estimator

Create a function that estimates delivery time based on distance, traffic level, and weather condition.

**Factors:**
- Distance (km)
- Traffic level
- Weather condition

**Real-Time Use:**
- Food delivery apps
- Logistics platforms

**Hint:**
- Use function + multiple conditions
- Apply mathematical adjustments

**Sample Input:**
Distance (km): 8  
Traffic Level: High  
Weather: Rainy  

**Expected Output:**
Estimated Delivery Time: 55 minutes

```python
# Function to estimate delivery time
def estimate_delivery_time(distance, traffic, weather):
    
    # Base time calculation (5 minutes per km)
    time = distance * 5
    
    # Traffic adjustment
    if traffic.lower() == "high":
        time += 10
    elif traffic.lower() == "medium":
        time += 5
    
    # Weather adjustment
    if weather.lower() == "rainy":
        time += 5
    elif weather.lower() == "stormy":
        time += 15
    
    return time

distance_km = 8
traffic_level = "High"
weather_condition = "Rainy"

eta = estimate_delivery_time(distance_km, traffic_level, weather_condition)

print("Estimated Delivery Time:", eta, "minutes")
```

## Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

Build a function that analyzes seat booking data and:
- Calculates occupancy percentage
- Determines if show is Housefull
- Suggests opening additional shows

**Real-Time Use:**
- Cinema ticketing systems
- Event management software

**Hint:**
- Function + list
- Percentage calculation

**Sample Input:**
Total Seats: 200  
Booked Seats: [1, 1, 1, ...] (150 entries)

**Expected Output:**
Occupancy: 75%
Show Status: Almost Full

```python
# Function to analyze seat occupancy
def analyze_occupancy(total_seats, booked_seats):
    
    booked_count = 0
    
    # Count booked seats using loop
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1
    
    occupancy_percentage = (booked_count / total_seats) * 100
    
    # Determine show status
    if occupancy_percentage == 100:
        status = "Seats Full"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    return occupancy_percentage, status

total_seats = 200
booked_seats = [1] * 150  # 150 seats booked

occupancy, show_status = analyze_occupancy(total_seats, booked_seats)

print("Occupancy:", int(occupancy), "%")
print("Show Status:", show_status)
```

## Problem Statement 4: Cloud Server Load Classification System

Create a function to classify server load based on CPU usage readings.

**Rules:**
- Average CPU < 50% --> Normal
- 50% – 80% --> Warning
- Above 80% --> Critical

**Real-Time Use:**
- Cloud monitoring dashboards
- DevOps alerting systems

**Hint:**
- Function + loop
- Average calculation

**Sample Input:**
CPU Readings (%): [45, 60, 70, 85, 90]

**Expected Output:**
Average CPU Load: 70%
Server Status: Warning

```python
# Function to classify server load
def classify_server_load(cpu_readings):
    
    total = 0
    
    # Calculate total CPU using loop
    for reading in cpu_readings:
        total += reading
    
    average_cpu = total / len(cpu_readings)
    
    # Classify server status
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    return average_cpu, status

cpu_data = [45, 60, 70, 85, 90, 80]

avg_cpu, server_status = classify_server_load(cpu_data)

print("Average CPU Load:", round(avg_cpu,2), "%")
print("Server Status:", server_status)
```

## Problem Statement 5: Smart Classroom Resource Usage Monitor

Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

**Real-Time Use:**
- Smart classrooms
- Energy optimization systems

**Hint:**
- Function + dictionary
- Conditional checks

**Sample Input:**
Resource Usage (hours):
{
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

**Expected Output:**
Overused Resources: AC
Energy Alert: Yes

```python
# Function to check classroom resource usage
def monitor_resources(resources):
    
    usage_limit = 8
    overused_resource = None
    
    # Loop through each resource
    for name in resources:
        
        # Check if usage exceeds limit
        if resources[name] > usage_limit:
            overused_resource = name
    
    if overused_resource:
        print("Overused Resources:", overused_resource)
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

## Problem Statement 6: Online Event Registration Capacity Controller

Create a function that manages event registrations by:
- Tracking registrations
- Preventing overbooking
- Triggering waitlist mode

**Real-Time Use:**
- Webinar platforms
- Conference registration systems

**Hint:**
- Function + loop
- Capacity validation

**Sample Input:**
Event Capacity: 100  
Registrations: 105  

**Expected Output:**
Confirmed Registrations: 100  
Waitlisted Users: 5  
Registration Status: Closed

```python
# Function to manage event registrations
def manage_registrations(capacity, registrations):
    
    confirmed = 0
    waitlisted = 0
    
    # Process each registration using loop
    for i in range(registrations):
        
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    # Check registration status
    if confirmed == capacity:
        status = "Closed"
    else:
        status = "Open"
    
    return confirmed, waitlisted, status

event_capacity = 100
total_registrations = 105

confirmed_users, waitlisted_users, registration_status = manage_registrations(
    event_capacity, total_registrations
)

print("Confirmed Registrations:", confirmed_users)
print("Waitlisted Users:", waitlisted_users)
print("Registration Status:", registration_status)
```
